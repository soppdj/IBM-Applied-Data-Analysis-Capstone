# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load data from the table named 'master'
query = "SELECT ConvertedComp FROM master"
df = pd.read_sql_query(query, conn)

# Drop NaN values from the 'ConvertedComp' column
df = df.dropna(subset=['ConvertedComp'])
```

```python
# Plot the histogram using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=50, edgecolor='k')
plt.title('Distribution of Converted Compensation')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load data from the table named 'master'
query = "SELECT Age FROM master"
df = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Drop NaN values from the 'Age' column
df = df.dropna(subset=['Age'])
```

```python
# Plot the box plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load data from the table named 'master'
query = "SELECT Age, WorkWeekHrs FROM master"
df = pd.read_sql_query(query, conn)

# Drop NaN values from the 'Age' and 'WorkWeekHrs' columns
df = df.dropna(subset=['Age', 'WorkWeekHrs'])
```

```python
# Plot the scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load data from the table named 'master'
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"
df = pd.read_sql_query(query, conn)

# Drop NaN values from the 'WorkWeekHrs', 'CodeRevHrs', and 'Age' columns
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])
```

```python
# Plot the bubble plot using Matplotlib
plt.figure(figsize=(10, 6))
bubble_size = df['Age'] / df['Age'].max() * 100  # Normalize bubble size
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs with Age as Bubble Size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load relevant columns from the 'master' table
query = "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master"
df = pd.read_sql_query(query, conn)

# Convert Age to numeric, coercing errors to NaN and drop rows with NaN values
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df = df.dropna(subset=['Age', 'WorkWeekHrs', 'CodeRevHrs'])
```

```python
# Filter the data for the age group 30 to 35
age_filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median WorkWeekHrs and CodeRevHrs for this age group
median_workweekhrs = age_filtered_df['WorkWeekHrs'].median()
median_coderevhrs = age_filtered_df['CodeRevHrs'].median()

# Create a dataframe for the medians
median_data = pd.DataFrame({
    'Metric': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median': [median_workweekhrs, median_coderevhrs]
})
```

```python
# Plot a stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Create the bar chart
bars = ax.bar(median_data['Metric'], median_data['Median'], color=['#1f77b4', '#ff7f0e'])

# Add text labels above the bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.1f}', 
                xy=(bar.get_x() + bar.get_width() / 2, height), 
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# Customize the chart
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
ax.set_ylabel('Median Hours')
ax.set_ylim(0, max(median_workweekhrs, median_coderevhrs) + 5)

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load relevant columns from the 'master' table
query = "SELECT Age, ConvertedComp FROM master"
df = pd.read_sql_query(query, conn)

# Convert Age and ConvertedComp to numeric, coercing errors to NaN and drop rows with NaN values
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['ConvertedComp'] = pd.to_numeric(df['ConvertedComp'], errors='coerce')
df = df.dropna(subset=['Age', 'ConvertedComp'])
```

```python
# Filter the data for the age range 45 to 60
age_filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age within this range
median_comp_by_age = age_filtered_df.groupby('Age')['ConvertedComp'].median().reset_index()
```

```python
# Plot a line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedComp'], marker='o', linestyle='-', color='b')

# Customize the chart
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True)
plt.xticks(median_comp_by_age['Age'])

# Display the chart
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load the 'MainBranch' column from the 'master' table
query = "SELECT MainBranch FROM master"
df = pd.read_sql_query(query, conn)

# Calculate the frequency of each unique value in the 'MainBranch' column
mainbranch_counts = df['MainBranch'].value_counts()
```

```python
# Plot a horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue')

# Customize the chart
plt.title('Distribution of Main Branch Responses')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.grid(axis='x')

# Display the chart
plt.show()
```
